```python
import pandas as pd
t2g2_main = pd.read_csv('t2g2_main.csv')
t2g2_imputed = pd.read_csv('t2g2_imputed.csv')
t1g2 = pd.read_csv('t1g2.csv')
t2g3 = pd.read_csv('t2g3.csv')
t2g5 = pd.read_csv('t2g5.csv')
amk = pd.read_csv('amk.csv')
device01 = pd.read_csv('device01.csv')
```

```python
t2g2_main.head()
```

```python
t2g2_main['device'] = ['T2G2_main' for x in t2g2_main.index]
t2g2_main.head()
```

```python
t2g2_imputed.head()
```

```python
t2g2_imputed['device'] = ['T2G2_imputed' for x in t2g2_imputed.index]
t2g2_imputed.head()
```

```python
t1g2.head()
```

```python
t1g2['device'] = ['T1G2' for x in t1g2.index]
t1g2.head()
```

```python
t2g3.head()
```

```python
t2g3['device'] = ['T2G3' for x in t2g3.index]
t2g3.head()
```

```python
t2g5.head()
```

```python
t2g5['device'] = ['T2G5' for x in t2g5.index]
t2g5.head()
```

```python
amk.head()
```

```python
amk.drop('index', axis=1, inplace=True)
amk['device'] = ['AMK' for x in amk.index]
amk.head()
```

```python
device01.head()
```

```python
device01['device'] = ['Device01' for x in device01.index]
device01.head()
```

```python
final = pd.concat([t2g2_main, t2g2_imputed, t1g2, t2g3, t2g5, amk, device01], axis=0)
final.head()
```

```python
final.fillna(0, inplace=True)
final.set_index('timestamp', inplace=True)
final.head()
```

```python
final.index = pd.to_datetime(final.index)
final.sort_index(inplace=True)
final.head()
```
